```python
from selenium import webdriver
from selenium.webdriver.common.by import By

##############################################
## Check the package details according to these lines
from selenium.webdriver.support.ui import WebDriverWait 
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import Select
##############################################

from datetime import datetime
import time
```

## Step 1. Define the functions to click button, fill in inputs, and skills 

```python
def click(selector: str, key: str, num):
        by_selector = getattr(By, dict_select[selector])
        button = driver.find_element(by_selector,key)
        for i in range(0, num):
            button.click()
```

```python
def fillc(selector: str, key: str, content: str):
        by_selector = getattr(By, dict_select[selector])
        driver.find_element(by_selector,key).clear()
        driver.find_element(by_selector,key).send_keys(content)
```

## Step 2. Define the initial parameters

```python
jobcode = '21321'
resume = 'D:/workshops/temp_World Bank Jobs Application Dashboard and Automate/CV and CL by job id codes/req' + jobcode + '_Shiyao Wang Resume.pdf'
coverletter ='D:/workshops/temp_World Bank Jobs Application Dashboard and Automate/CV and CL by job id codes/req' + jobcode + '_Shiyao Wang Cover Letter.pdf'

skillsets = {
    
    'econ': {
        'include': True,
        'list': ['CFA','Project Management','Economic Modeling', 'Balance of Payments', 'Macro-Economic Research', 'Economic Data Pre-Processing', 'Exchange Rate Assessment and Surveillance', 'Economic Report Writing']
        },
    
    'datapro': {
        'include': True,
        'list': ['Stata', 'Python', 'Excel VBA', 'R', 'SQL']
        },
    
    # add more skillsets as needed
    
}

dict_select = {"uid" : "ID",
               "css" : "CSS_SELECTOR",
               "xpa" : "XPATH",
               "link" : "LINK_TEXT",
               "tag" : "TAG_NAME"
              }

explist = {
    'imf': {
        'title': 'Research Analyst',
        'org': 'International Monetary Fund',
        'position': 'Washington D.C.',
        'start': '07/2019',
        'end': '' 
        },
    
    'world bank': {
        'title': 'Consultant',
        'org': 'World Bank Group',
        'position': 'Washington D.C.',
        'start': '08/2018',
        'end': '05/2019' 
        },
    
    'jhu': {
        'title': 'Teaching Assistant',
        'org': 'Carey Business School',
        'position': 'Washington D.C.',
        'start': '08/2018',
        'end': '05/2019' 
        },
    
    'gu': {
        'title': 'Teaching Assistant',
        'org': 'Georgetown University',
        'position': 'Washington D.C.',
        'start': '09/2017',
        'end': '05/2018' 
        }
}

edulist = {
    'public policy': {
        'level': 'Masters Level (ML)',
        'institution': 'Georgetown University',
        'country': 'United States (USA)',
        'earned': '06/2016',
        'name': 'Public Policy',
        'field': 'Econometrics' ## This may be modified
        },
    
    'business administration': {
        'level': 'Bachelors Level (BL)',
        'institution': 'Wuhan University',
        'country': 'China (CHN)',
        'earned': '06/2019',
        'name': 'Business Administration',
        'field': 'Management' ## This may be modified
        },
    
    'english literature': {
        'level': 'Bachelors Level (BL)',
        'institution': 'Wuhan University',
        'country': 'China (CHN)',
        'earned': '06/2019',
        'name': 'Arts',
        'field': 'English Literature' ## This may be modified
        }
}
```

## Step 3. Start the Engine and Fill in!

```python
global driver
driver = webdriver.Chrome()
driver.get('https://worldbankgroup.csod.com/ux/ats/careersite/1/requisition/' + jobcode + '/application?c=worldbankgroup#2')
time.sleep(3)


##### Step 2 of 9 
#######################################
# Page Step 1: Login and go to Qualifications and etc. page
#######################################
fillc('uid', 'actionItem.firstName.idTag', 'Shiyao') # used the pre-defined function to fill in input boxes
fillc('uid', 'actionItem.lastName.idTag', 'Wang')
fillc('uid', 'actionItem.email.idTag', 'sw1016@georgetown.edu')
click('css', 'button[data-tag="btnNext"]', 1)
time.sleep(3)

click('css', 'button[data-tag="loginDialogLoginBtn"]',1)
time.sleep(3)

fillc('css', 'input[data-tag="txtPassword"]', '20130301Math!')
click('css', 'a[data-tag="btnSignIn"]', 1)
time.sleep(3)

click('css', 'button[data-tag="btnNext"]', 1)
time.sleep(3)

# Page Step 2: Remove the resume and upload a new one
#######################################
click('xpa', '//div[@data-tag="resumeFileViewer"]//*[contains(@title, "Delete File")]', 1)
time.sleep(1)

driver.find_element(By.XPATH,"//label[@for='resumeFileUpload']//input[@type='file']").send_keys(resume)
time.sleep(10)

# Page Step 3: Contact Details
#######################################
# fillc('uid', 'contactDetails_addressLine1', '1643 S Hayes Street Apt B')
# fillc('uid', 'contactDetails_city', 'Arlington')
# fillc('uid', 'contactDetails_state', 'Virginia')
# Select(driver.find_element(By.ID, 'contactDetails_country')).select_by_visible_text('United States')
# fillc('uid', 'contactDetails_zipCode', '22202')
fillc('uid', 'contactDetails_phone', '2027021004')

# Page Step 4: Experience
#######################################
try:
    click('link', 'Delete Experience', 10)
except:
    click('link', 'Add Experience', 3)

for i in range(0,4):
    for j in range(0,5):
        cell = '$resume-field_0_' + str(i) + '_' + str(j)
        inputvalue = explist[list(explist)[i]][list(explist[list(explist)[i]])[j]]
        fillc('uid', cell, inputvalue)

# Page Step 5: Education
#######################################
try:
    click('link', 'Delete Education', 10)
except:
    click('link', 'Add Education', 2)

for i in range(0,3):
    for j in range(0,6):
        cell = '$resume-field_1_' + str(i) + '_' + str(j)
        inputvalue = edulist[list(edulist)[i]][list(edulist[list(edulist)[i]])[j]]
        if j in [0,2]:
            Select(driver.find_element(By.ID, cell)).select_by_visible_text(inputvalue)
        else:
            fillc('uid', cell, inputvalue)
        
# Page Step 6: Skill/Certification/Language
#######################################
try:
    click('link', 'Delete Skill / Certification / Language', 30)
except:
    pass

y = 0
prevnum = 0
for skill in skillsets:
    if skillsets[skill]['include'] == True:
        click('link', 'Add Skill / Certification / Language', len(skillsets[skill]['list']))
        global y
        x = y
        y = y + len(skillsets[skill]['list'])
        for i in range(0,len(skillsets[skill]['list'])):
            field = '$resume-field_2_' + str(i+prevnum) + '_0'
            field_level = '$resume-field_2_' + str(i+prevnum) + '_1'
            driver.find_element(By.ID,field).clear()
            driver.find_element(By.ID,field).send_keys(skillsets[skill]['list'][i])
            Select(driver.find_element(By.ID,field_level)).select_by_visible_text('1 - High')
        prevnum += y  
    else:
        print(skillsets[skill] +  'skill not included')

driver.find_elements(By.LINK_TEXT,'Delete Skill / Certification / Language')[-1].click()

# Page Step 7: MANYDATORY Personal Information
#######################################
Select(driver.find_element(By.ID, '$resume-field_3_0_0')).select_by_visible_text('Male')
fillc('uid', '$resume-field_3_0_1', '6/2/1994')
Select(driver.find_element(By.ID, '$resume-field_3_0_2')).select_by_visible_text("China - People's Republic of China")

# Page Step 8: Cover Letter
#######################################
driver.find_element(By.XPATH,"//label[@for='cover_letter_file_upload']//input[@type='file']").send_keys(coverletter)
time.sleep(3)

# Page Step 9: Acknowledgement
#######################################
click('xpa', '//label[@for="user-agreement-374985"]', 1)
time.sleep(1)

click('css', 'button[data-tag="btnNext"]', 1)
time.sleep(3)


##### Step 3 of 9 
#######################################
driver.find_element(By.XPATH, "//label[contains(text(), 'Former WBG Consultant/Temp')]").click()
driver.find_element(By.XPATH, "//label[contains(text(), 'International Organization')]").click()
driver.find_element(By.XPATH, "//label[contains(text(), 'Masters Level (ML)')]").click()
relatives = driver.find_element(By.XPATH, "//div[contains(text(), 'Do you have any relatives working for World Bank Group')]")
relatives.find_element(By.XPATH, "./following-sibling::div//label[contains(text(), 'No')]").click()
driver.find_element(By.XPATH, "//label[contains(text(), 'Male')]").click()
driver.find_element(By.XPATH, "//div[contains(text(), 'What is your date of birth')]").click()
birth = driver.find_element(By.XPATH, "//div[contains(text(), 'What is your date of birth')]")
birth.find_element(By.XPATH, "./following-sibling::div//input").send_keys('6/2/1994')
time.sleep(1)

click('css', 'button[data-tag="btnNext"]', 1)
time.sleep(3)


##### Step 4 of 9 
#######################################
driver.find_element(By.XPATH, "//label[contains(text(), 'China - People')]").click()
click('css', 'button[data-tag="btnNext"]', 1)
time.sleep(3)


##### Step 5 of 9 
#######################################
driver.find_element(By.XPATH, "//label[contains(text(), 'Not applicable')]").click()
click('css', 'button[data-tag="btnNext"]', 1)
time.sleep(3)


##### Step 6 of 9 
#######################################
driver.find_element(By.XPATH, "//label[contains(text(), 'Not applicable')]").click()
click('css', 'button[data-tag="btnNext"]', 1)
time.sleep(3)


##### Step 7 of 9 
#######################################
# This will automatically select all the first choices
divs = driver.find_elements(By.XPATH, "//div[contains(text(), '?')]") # find all the questions
for i in divs:
    i.find_elements(By.XPATH, "./following-sibling::div//label")[0].click()
click('css', 'button[data-tag="btnNext"]', 1)
time.sleep(3)


##### Step 8 of 9 
#######################################
driver.find_element(By.XPATH, "//label[contains(text(), 'Through Current World Bank/IFC Employee')]").click()
fillc('tag', 'textarea', 'Yuwei Wang')
```

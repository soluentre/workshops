```python
import requests
import gzip
import pandas as pd
from io import StringIO

url = 'https://comtradeapi.un.org/bulk/v1/file/52/863302c6d3c6da81ce650ac543479a1743a4f1b57a98187e630a3bafbfcf7fb5'
headers = {'Ocp-Apim-Subscription-Key': '78f8526e63744385beacee6f0d3f2ac5'}

response = requests.get(url, headers=headers)

# Decode the content using gzip
content = gzip.decompress(response.content)

# Decode the content using the appropriate encoding
content = content.decode('utf-8')
df = pd.read_csv(StringIO(content), sep='\t')
```
